```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Carregar a imagem
imagem_colorida = cv2.imread('flor.png')  # Substitua 'lena.png' pelo nome do arquivo
imagem_colorida = cv2.cvtColor(imagem_colorida, cv2.COLOR_BGR2RGB)  # Converter de BGR para RGB
```

```python
# Converter para níveis de cinza
imagem_cinza = cv2.cvtColor(imagem_colorida, cv2.COLOR_RGB2GRAY)
```

```python
# Binarizar a imagem
_, imagem_binarizada = cv2.threshold(imagem_cinza, 127, 255, cv2.THRESH_BINARY)
```

```python
# Exibir as imagens
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
plt.imshow(imagem_colorida)
plt.title('Imagem Colorida')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(imagem_cinza, cmap='gray')
plt.title('Imagem em Níveis de Cinza')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(imagem_binarizada, cmap='gray')
plt.title('Imagem Binarizada (Preto e Branco)')
plt.axis('off')

plt.tight_layout()
plt.show()
```
