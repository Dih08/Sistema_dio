```python
!pip install mtcnn
import cv2
import numpy as np
from mtcnn.mtcnn import MTCNN

# Inicializar o detector MTCNN
detector = MTCNN()

# Função para detectar faces em uma imagem
def detect_faces(image):
    faces = detector.detect_faces(image)
    return faces

# Carregar uma imagem
image = cv2.imread('imagem24.jpg')
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Detectar faces
faces = detect_faces(image_rgb)

# Desenhar caixas delimitadoras ao redor das faces detectadas
for face in faces:
    x, y, width, height = face['box']
    cv2.rectangle(image_rgb, (x, y), (x + width, y + height), (255, 0, 0), 2)

# Exibir a imagem
import matplotlib.pyplot as plt
plt.imshow(image_rgb)
plt.axis('off')
plt.show()
```
